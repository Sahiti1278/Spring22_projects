Question 1: 

The attached carprices.txt contains a single column of car prices in 1000â€™s. Read this data into your program. 

We load this data using pandas

```python
#The first line is the heading and should be ignored. 
data = pd.read_csv("carprices.txt",header=0)
#For any line containing non-numeric data, raise a ValueError exception and print a message saying â€˜This data is invalid and will be ignoredâ€™ together with the exception object.
try:
  int(data)
except :
    print ('This data is invalid and will be ignored')
data = data.apply(pd.to_numeric, errors='coerce').dropna()
#You should divide each number by 10 and round it down to the nearest integer. 
data_div = data/10
data_round = round(data_div)
#Then plot a histogram that looks like the one shown below. The number of bins to use is 50.
plt.hist(data_round.to_numpy(),bins=50,rwidth=0.85)
plt.title('Frenquency of Car Prices', fontsize= 15)
plt.xlabel('Price in tens of thousands', fontsize= 15)
plt.ylabel('Number of cars', fontsize= 15)
plt.xlim((0,50))
plt.ylim((0,10))
plt.tight_layout(pad=0.05, w_pad=0.05, h_pad=0.05)
plt.show() 
plt.close()
```

Question 2:

```python
#The first line is the heading and should be ignored. 
data = np.loadtxt("Wildlife.txt", skiprows=1,dtype=int)#ignore 1st line here
#print(data)
year = data[:,0]
Bears = data[:,1]
Dolphins = data[:,2]
Whales = data[:,3]
plt.plot(year,Bears,'b',label = 'Bears')
plt.plot(year,Dolphins,color='orange', label = 'Dolphins')
plt.plot(year,Whales,'g',label = 'Whales')
plt.legend(bbox_to_anchor=(0.24, 0.98), loc=1, borderaxespad=0)
plt.xlabel('Years', fontsize= 15)
plt.ylabel('Wildlife Population', fontsize= 15)
plt.title('Wildlife Population Per Year', fontsize= 15)
plt.tight_layout(pad=0.05, w_pad=0.05, h_pad=0.05)
plt.show()
plt.close()
```

```python
#The first line is the heading and should be ignored. 
data = pd.read_csv("cartype.txt",header=0)
# output = data.value_counts()
# ax = output.plot(kind='pie',autopct='%1.0f%%',ylabel='',subplots=True, label=True)
labels = data['Type'].unique().tolist()
labels.sort() #sort alphabatically
output = data.value_counts().sort_index(ascending=True) #sort the type in alphabatic order
#print(labels, counts)
explode = (0.05, 0.05, 0.05, 0.05, 0.05, 0.05)
plt.pie(output, labels=labels, explode=explode, autopct='%1.2f%%',startangle=90, textprops={'fontsize': 14})#, explode=explode, radius = 2)
plt.tight_layout(pad=0.05, w_pad=0.05, h_pad=0.05)
plt.axis('equal')  #taken from single circle
plt.show()
plt.close()
```
